### 1. A stiff system of ODEâ€™s
The following shows 2 graphs. One where N is not large enough for RK2 to be stable and the other where $N$ is large enough to the point that the difference between it and odeint is neglegible.

```python
from numpy import pi,sin,linspace,sqrt,cos
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

def dudt(u,v,t):
    return 998*u + 1998*v

def dvdt(u,v,t):
    return -999*u - 1999*v

def f(uv,t):
    u,v = uv
    
    dudt = 998*u + 1998*v
    dvdt = -999*u - 1999*v
    
    return dudt,dvdt

def rk2(dudtdvdt, u0v0, t):
    dudt,dvdt = dudtdvdt
    u_0,v_0 = u0v0
    
    len_t = len(t)
    u = np.empty(len_t)
    v = np.empty(len_t)
    
    u[0] = u_0
    v[0] = v_0
    
    h = t[1]-t[0]
    
    for i in range(1, len(t)):
        k1 = h*dudt(u[i-1], v[i-1], t[i-1])
        l1 = h*dvdt(u[i-1], v[i-1], t[i-1])
        k2 = h*dudt(u[i-1] + k1/2, v[i-1] + l1/2, t[i-1] + h/2)
        l2 = h*dvdt(u[i-1] + k1/2, v[i-1] + l1/2, t[i-1] + h/2)
        
        u[i] = u[i-1] + k2
        v[i] = v[i-1] + l2
        
    return u,v

for i in range(3,5):
    t = linspace(0,3,10**i)
    u_0 = 2
    v_0 = -1

    uv_o = odeint(f,(u_0,v_0),t)
    uv_r = rk2((dudt,dvdt),(u_0,v_0),t)
    
    plt.title(f'$N = 10^{i}$')
    plt.ylim(-2,4)
    plt.plot(t,uv_o[:,0],label='$u(t)$:odeint')
    plt.plot(t,uv_o[:,1],label='$v(t)$:odeint')
    plt.xlabel('$t$')
    plt.ylabel('$u$ or $v$')

    plt.plot(t,uv_r[0],label='$u(t)$:rk2')
    plt.plot(t,uv_r[1],label='$v(t)$:rk2')
    plt.xlabel('$t$')
    plt.ylabel('$u or v$')

    plt.legend()
    plt.show()
```

### 2. The nonlinear pendulum
I solved the differential equation like the same methods as before and $\theta(t)$ is a sine wave.
$$T = 2\pi\sqrt{\frac{l}{g}} = 2\pi\sqrt{\frac{1}{g}} \approx 2\space s$$
The calculated period concurs with what we observe in the graph.

```python
m = 1 #Kg
l = 1 #m
g = 9.8 #m/s^2

def f(thetaw,t):
    theta,w = thetaw
    
    dthetadt = w
    dwdt = -g/l * sin(theta)
    
    return dthetadt,dwdt

t = linspace(0,5,1000)
theta_0 = 0
w_0 = 1 #rad/s

thetaw = odeint(f,(theta_0,w_0),t)

plt.plot(t,thetaw[:,0])
plt.xlabel('$t$ (s)')
plt.ylabel(r'$\theta$ (rad)')
plt.show()
```

#### b)
$\omega_0 = 6.122448979591837$ (next $\omega_0$ must be larger than the critical value) is the last value for which $\theta(t)$ is still sinusoidal. For large $\omega_0$. the initial angular velocity causes an initial Kinetic energy which is greater than the maximum Potential energy. This means that the angular velocity at the highest point is non zero. Since there is no damping term, it keeps going around, ad inifinitum.

$$K = \frac{1}{2}mv^2 = \frac{1}{2}ml^2 \omega^2$$ (Note: $v = l \omega$)
$$U_{max} = 2mgl$$

Setting them equal to each other since the behavior changes at $K = U_{max}$.

$$\frac{1}{2}ml^2 \omega^2 = 2mgl \implies \omega = 4\frac{g}{l} \implies \omega = 2\sqrt{\frac{g}{l}}$$
$$\omega_{critical} = 2\sqrt{\frac{g}{l}} = 2\sqrt{g} \approx 6.26$$

```python
plt.title('50 different $\omega_0$ values')

thetaws = [odeint(f,(theta_0,w_0),t) for w_0 in linspace(0,10,50)]

for thetaw in thetaws:
    plt.plot(t,thetaw[:,0])
    plt.xlabel('$t$ (s)')
    plt.ylabel(r'$\theta$ (rad)')
    
plt.show()
```

#### c)
The trajectories in the phase space that form closed surves correspond to those in part b) that are sinusoidal. Note that the curves that are oscilating  in phase space are the ones that correspond to the surves going off to infinity in part b).

```python
plt.xlim(-pi,pi)

for thetaw in thetaws:
    plt.plot(thetaw[:,0],thetaw[:,1])
    plt.xlabel(r'$\theta$ (rad)')
    plt.ylabel('$\omega_0$ (rad/s)')
    
plt.show()
```

### 3. Newman Ex. 8.7: Cannonball with air resistance

Using
$$F_g = -g\hat j$$
$$F_D = \frac{1}{2} \pi R^2 \rho C v^2$$
$$\sum\vec F = m \vec a \implies F_g - F_D = m\frac{d^2\vec r}{dt^2}$$
We get
$$\ddot x =  -\frac{\pi R^2\rho C}{2m}\cos(\theta)v^2
\implies \ddot x = -\frac{\pi R^2\rho C}{2m}\frac{\dot x}{\sqrt{\dot x^2 + \dot y^2}}(\dot x^2 + \dot y^2)
\implies \ddot x = -\frac{\pi R^2\rho C}{2m}\dot x \sqrt{\dot x^2 + \dot y^2}$$

$$\ddot y = -g -\frac{\pi R^2\rho C}{2m}\sin(\theta)v^2
\implies \ddot y =-g -\frac{\pi R^2\rho C}{2m}\frac{\dot y}{\sqrt{\dot x^2 + \dot y^2}}(\dot x^2 + \dot y^2)
\implies \ddot y =-g -\frac{\pi R^2\rho C}{2m}\dot y \sqrt{\dot x^2 + \dot y^2}$$

#### a)

```python
m = 1 #Kg
r = .08 #m
p = 1.22 #Kgm^-3
C = .47

def f(vec,t):
    x,v_x,y,v_y = vec
    
    dxdt = v_x
    dydt = v_y
    
    const = -pi*(r**2)*p*C/(2*m) * sqrt(v_x**2 + v_y**2)
    
    dv_xdt = const*v_x
    dv_ydt = -g + const*v_y
    
    return dxdt,dv_xdt,dydt,dv_ydt

v_x0 = 100*cos(pi/6)
v_y0 = 100*sin(pi/6)

t = linspace(0,7,10000)
xvxyvy = odeint(f,(0,v_x0,0,v_y0),t)

plt.ylim(0,60)
plt.title('$y$ vs $x$')
plt.plot(xvxyvy[:,0],xvxyvy[:,2])
plt.xlabel('$x$ (m)')
plt.ylabel('$y$ (m)')

plt.show()
```

#### b)
Looping over a range of mass values, we can see that x approaches some value around 825.

$$\lim_{m\to\infty}\ddot x = 0$$
$$\lim_{m\to\infty}\ddot y = -g$$

So, as mass approaches infinity, the trajectory approaches a simple parabola (where the air resistance is neglegible).

```python
t = linspace(0,100,10000)
plt.xlim(0,1000)
plt.ylim(0,150)

for m in linspace(.1,100,100):
    xvxyvy = odeint(f,(0,v_x0,0,v_y0),t)

    plt.title('$y$ vs $x$')
    plt.plot(xvxyvy[:,0],xvxyvy[:,2])
    plt.xlabel('$x$ (m)')
    plt.ylabel('$y$ (m)')
    
plt.show()
```

### 4. Newman Ex. 8.3: The Lorenz equations (deterministic chaos)
Despite starting out the same, the two graphs become drastically different despite only a difference of $10^{-9}$ in $y_0$.

```python
s = 10
r = 28
b = 8/3

def f(vec,t):
    x,y,z = vec
    
    dxdt = s*(y-x)
    dydt = r*x - y - x*z
    dzdt = x*y - b*z
    
    return dxdt,dydt,dzdt

t = linspace(0,50,10000)

for y_0 in [1,1+1e-9]:
    xyz = odeint(f,(0,y_0,0),t)

    plt.xlabel('$t$')
    plt.ylabel('$y(t)$')
    plt.plot(t,xyz[:,1],label=f'$y_0 = {y_0}$')

plt.title('Deterministic Chaos')
plt.legend()
plt.show()
```

#### b)
Plotting $z$ against $x$ using $y_0 = 0$ gives us the following graph.

```python
xyz = odeint(f,(0,1,0),t)

x = xyz[:,0]
y = xyz[:,1]
z = xyz[:,2]

plt.title('$z$ vs $x$')
plt.xlabel('$z$')
plt.ylabel('$x$')
plt.plot(x,z)

plt.show()
```

### 5. 3D plot of the Lorenz strange attractor
Plotting the obtained $x$,$y$, and $z$ values in a 3D plot gives us the following beautiful graph.

```python
from mpl_toolkits.mplot3d import Axes3D

plt.figure(figsize=(10,10))
plt.rc('font', size=12)
ax = plt.gca(projection='3d')
ax.plot(x,y,z,c='m')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')
plt.title('Lorenz strange attractor')
plt.show()
```
