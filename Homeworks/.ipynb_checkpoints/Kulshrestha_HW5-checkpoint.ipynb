### 1. Computing Bessel functions
I imported the trapezoidal integration from one of the activities and used it, and the quad function to integrate the desired function (defined as an lambda function). I then plotted it alongside the imported $J_1$ function.

```python
import numpy as np
from numpy import linspace,cos,sin,pi,exp,sqrt
from scipy.special import j1
from scipy.integrate import quad
import matplotlib.pyplot as plt

def trap(f,low,high,N,x):
    ls = linspace(low,high,N)
    coeff = np.full(N,1)
    coeff[0] = .5
    coeff[-1] = .5
    
    return np.array([((high-low)/N) * np.sum(coeff*f(ls,xi)) for xi in x])

N = 1000
x = linspace(0,20,N)

f = lambda u,x: cos(u - x*sin(u))

a = j1(x)
b = 1/pi * trap(f,0,pi,N,x)
c = 1/pi * np.array([quad(f,0,pi,args=xi)[0] for xi in x])

plt.title('Bessel functions')
plt.ylabel('$J_1(x)$')
plt.xlabel('x')
plt.plot(x,a+.1,label='scipy.special: offset = +0.1')
plt.plot(x,b,label='trapezoid rule')
plt.plot(x,c-.1,label='scipy.integrate.quad: offset = -0.1')
plt.legend()

plt.show()
```

### 2. Newman Ex. 5.4(b): The diffraction limit of a telescope
I defined a function for $I(r)$ uisng the formula provided by Newman which is used in a double for loop iterating over a double array (as opposed to vectorizing for 2 dimensions). This information is then plotted using imshow().

```python
l = 500e-9 #nm

def j1_(x):
    return quad(f,0,pi,args=x)[0]

def I(r):
    kr = 2*pi/l * r
    return (j1_(kr)/kr)**2

plt.figure(figsize=(9,6))

N = 500
lim = 1e-6
x = linspace(-lim,lim,N)
y = linspace(-lim,lim,N)
Ir = np.empty((N,N))

n_range = range(N)
for i in n_range:
    for j in n_range:
        r = sqrt((x[i])**2 + (y[j])**2)
        Ir[i,j] = I(r)

plt.title('Diffraction limit of a telescope')
plt.xlabel('$x (micrometers)$')
plt.ylabel('$y (micrometers)$')
plt.imshow(Ir,origin='lower',extent=(-1,1,-1,1),cmap='hot')
plt.colorbar()

plt.show()
```

### 3. Newman Ex. 5.9: Heat capacity of a solid
I implemented a function for $C_V$ using the formula provided by Newman and made it capable of handling an array of values and return a plottable numpy array.

```python
V = 1 #m^3
p = 6.022e28 #m^-3
t_D = 428 #K
k_B = 1.380e-23 #J/s
N = 50

def cv(T):
    b = t_D/T
    func = lambda x: (x**4 * exp(x))/((exp(x)-1)**2)
    cons = 9*V*p*k_B
    
    return np.array([cons*((1/bi)**3)*quad(func,0,bi)[0] for bi in b])

T = linspace(5,500,50)

plt.title('Heat capacity of Aluminum')
plt.xlabel('T')
plt.ylabel('$C_V$')
plt.plot(T,cv(T))
plt.show()
```
