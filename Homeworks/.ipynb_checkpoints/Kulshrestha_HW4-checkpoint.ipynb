### 1. Newman Ex. 6.1: A circuit of resistors

#### a)
The four equations are:
$$\frac{V_1 - V_2}{R} + \frac{V_1 - V_3}{R} + \frac{V_1 - V_4}{R} + \frac{V_1 - V_+}{R} = 0 \implies 4V_1 - V_2 - V_3 - V_4 = V_+$$

$$\frac{V_2 - V_1}{R} + \frac{V_2 - V_4}{R} + \frac{V_2 - 0}{R} = 0 \implies -V_1 + 3V_2 - V_4 = 0$$

$$\frac{V_3 - V_+}{R} + \frac{V_3 - V_1}{R} + \frac{V_3 - V_4}{R} = 0 \implies -V_1 + 3V_3 - V_4 = V_+ $$

$$\frac{V_4 - V_3}{R} + \frac{V_4 - V_1}{R} + \frac{V_4 - V_2}{R} + \frac{V_4 - 0}{R} = 0 \implies -V_1 - V_2 - V_3 + 4V_4 = 0$$

#### b)
This system of equations can be represented as a matrix $(A \space b)$:

\begin{pmatrix} 
4 & -1 & -1 & -1 & 5 \\
-1 & 3 & 0 & -1 & 0 \\
-1 & 0 & 3 & -1 & 5 \\
-1 & -1 & -1 & 4 & 0 \\
\end{pmatrix}

```python
import numpy as np
from scipy import linalg as la

A = np.array(
    [[4,-1,-1,-1],
    [-1,3,0,-1],
    [-1,0,3,-1],
    [-1,-1,-1,4]]
)

b = np.array(
    [[5],
    [0],
    [5],
    [0]]
)

V = la.solve(A,b)

for i, v in enumerate(V, start=1):
    print(f'V_{i} = {v[0]:.3f}')
```

### 2. Rotate the galaxy

Implemented a function that returns an N (given N = 3) dimensional rotational matrix, given N, the angle, and axis of rotation. This matrix is then multiplied with the data for the desired outcome. This data is then plotted in a scatter plot similarly to how it is in one of the previous handouts.

In this example, the galaxy has been rotated $\frac{\pi}{2}$ degrees around the axis specfied by the vector $\begin{pmatrix}
    1\\
    1\\
    0
\end{pmatrix}$

```python
""""Read file with coordinates of many points,
them make 3D scatter plot.
"""
from numpy import pi,cross,linspace
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def rmat(n,theta,nhat):
    """Returns 3D rotation matrix thru theta(rad) about unit vector nhat"""
    return la.expm(cross(np.eye(n),nhat*theta))

# Read the data from file:
data = np.loadtxt('../cpresources/galaxy2.txt',float,skiprows=1)

x = data[:,0]
y = data[:,1]
z = data[:,2]

plt.figure(figsize=(10,10))
plt.rc('font', size=12)
ax = plt.gca(projection='3d')
ax.scatter(x,y,z,s=1,c='m')
psize = 25.0
ax.set_xlim(-psize, psize)
ax.set_ylim(-psize, psize)
ax.set_zlim(-psize, psize)
ax.set_xlabel('x (kpc)')
ax.set_ylabel('y (kpc)')
ax.set_zlabel('z (kpc)')
plt.title(f'Imaginary Galaxy with {len(x)} stars before rotation')
plt.show()

rotation = np.array([1,1,0])
rotation = rmat(3,pi/2,rotation/la.norm(rotation))
data = data@rotation
print(data.shape)
# data = np.matmul(data, rotation) # does the same thing

x = data[:,0]
y = data[:,1]
z = data[:,2]

plt.figure(figsize=(10,10))
plt.rc('font', size=12)
ax = plt.gca(projection='3d')
ax.scatter(x,y,z,s=1,c='m')
psize = 25.0
ax.set_xlim(-psize, psize)
ax.set_ylim(-psize, psize)
ax.set_zlim(-psize, psize)
ax.set_xlabel('x (kpc)')
ax.set_ylabel('y (kpc)')
ax.set_zlabel('z (kpc)')
plt.title(f'Imaginary Galaxy with {len(x)} stars after rotation')

plt.show()
```

### 3. Use complex numbers to analyze an AC circuit
f is defined over the specified interval in Hz. I noticed that np.complex was not compatible with vectorizing, so I created my own version of it using np.vectorize for me to use. I then plotted the real axis with the imaginary axis against the frequency in kHz. We can see that there is a big peak in Re[I]  at $f_0 = \frac{1}{2\pi\sqrt{LC}} = 119\;kHz$

Note that I've defined a custom plot function for convenience.

```python
f = linspace(10e3, 250e3, 100) # Hz
w = 2*pi*f

R = 33 #ohms
L = 150e-6 #H
C = 12e-9 #F
V = 30 #V

complex_vectorized = np.vectorize(np.complex);

Z_L = complex_vectorized(0, w*L)
Z_C = 1/complex_vectorized(0, w*C)

Z_t = R+Z_L+Z_C

I = V/Z_t

def plot(x, ys, title, x_label, y_label, x_lim=None, y_lim=None, fig_size=None, plot_params=None, sub_plot=None, show=True):
    if fig_size:
        plt.figure(figsize=fig_size)
    if sub_plot:
        plt.subplot(sub_plot)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    if isinstance(ys, tuple):
        for y in ys:
            if plot_params:
                plt.plot(x, y, *plot_params)
            else:
                plt.plot(x, y)

    else:
        if plot_params:
            plt.plot(x, ys, *plot_params)
        else:
            plt.plot(x, ys)
    if x_lim:
        plt.xlim(*x_lim)
    if y_lim:        
        plt.ylim(*y_lim)
    if show:
        plt.show()
        
plot(f/1e3, (I.real, I.imag), 'AC circuit', 'f (kHz)', 'I (A)')
```

### 4. Using linalg.solve for an AC circuit
Using Kirchhoff's Voltage rule, we get the following equations:

$$ \frac{V_a-V}{R_1} + \frac{V_a-V_b}{Z_C} + \frac{V_a}{Z_L} = 0 $$
$$ \frac{V_b-V}{R_2} + \frac{V_b-V_b}{Z_C} + \frac{V_b}{R_3} = 0 $$

Which can be expressed and solved using matrices to obtain $V_a$ and $V_b$.

```python
def get_V(f):
    w = 2*pi*f
    R1 = 330 # Ohms
    R2 = 10
    R3 = 22
    L = 150e-6 # Henrys
    C = 12e-9 # F
    V = 30 # V
    
    Z_L = complex_vectorized(0, w*L)
    Z_C = 1/complex_vectorized(0, w*C)
    
    A = np.array([[(1/Z_L + 1/Z_C + 1/R1),    (-1/Z_C)],
                  [(-1/Z_C),    (1/R3 + 1/Z_C + 1/R2)]])
    b = np.array([V/R1, V/R2])
    
    return la.solve(A, b)

for _ in range(2):
    f = float(input('f (kHz)? '))*1e3 #Hz
    Va, Vb = get_V(f)
    print('\tVa = ', Va)
    print('\tVb = ', Vb)
```

### 5. Plot $V_a(f)$ and $V_b(f)$
Defined the frequency interval and plotting gives us:

```python
f = linspace(10e3, 250e3, 100) #Hz
Va = np.zeros(100,dtype=complex)
Vb = np.zeros(100,dtype=complex)

for i, freq in enumerate(f):
    Va[i], Vb[i] = get_V(freq)

plot(f/1e3, (Va.real, Va.imag), '$V_a(f)$ v f', 'f (kHz)', '$V_a(f)$', fig_size=(12,5), sub_plot=121, show=False)
plot(f/1e3, (Vb.real, Vb.imag), '$V_b(f)$ v f', 'f (kHz)', '$V_b(f)$', sub_plot=122)
```

#### c)
Making a second array that's twice as long (as per the hints) to allow curve_fit to deal with both parts of y (real and imaginary) at the same time nets the following:

```python
from scipy.optimize import curve_fit

plot(x, y, 'High Energy Physics Experiement', 'x', 'f(x)', plot_params='.', show=False)

x2 = np.concatenate((x,x))
y2 = np.concatenate((y[0],y[1]))
N = len(x)

def f2(x2, A, x_0, w):
    '''Model function for curve_fit, works with double-length real arrays x2, y2.'''
    x = x2[:N]
    y = f(x, A, x_0, w)
    return np.concatenate((y.real, y.imag))

params, C = curve_fit(f2, x2, y2, p0=guesses)

curve = f2(x2, *params)

plot(x2[:N], curve[:N], '', '', '', show=False)
plot(x2[N:], curve[N:], 'Using curve fit', 'x', 'f(x)')
```
